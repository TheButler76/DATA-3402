### Actual Python code

```python
f=open("Scores.csv", "r") 
# r=read only.
# w=write a new file.
# a=append.
# x=exlusive creation. creates a new file, but fails if the file already exists.
# b=binary files. read from binary files.

first_line=f.readline() # read the first line from our file. Which are also our columns!
print(first_line)

line=f.readline() 
while line:
    print(line)
    line=f.readline()
f.close() # We need to close the f function in order to read our data file.
```

```python
f=open('Scores.csv', 'r')
first_line=f.readline()
print(first_line.split(","))
f.close()
# We notice that there is a newline character at the end of the list...
```

```python
f=open('Scores.csv', 'r')
# .rstrip() removes the last newline character.
first_line=f.readline().rstrip()
print(first_line.split(","))
f.close()
```

```python
f=open('Scores.csv', 'r')
first_line=f.readline().rstrip()
keys=first_line.split(',')
print(keys)
f.close()
```

```python
f=open('Scores.csv', 'r')
first_line=f.readline().rstrip()
keys=first_line.split(',')

data=list()
line=f.readline().rstrip()
while line:
    data.append(line.split(","))
    line=f.readline().rstrip()
    print(line)
    
f.close()
```

### Indexing

```python
class my_list:
    def __init__(self, a_list):
        self._list=a_list
        
    def __getitem__(self, key):
        print(key)
        pass
```

```python
obj=my_list([5,5,5])
obj[1]
```

```python
obj[1,2]
obj[1,2,3]
```

```python
class my_list:
    def __init__(self, a_list):
        self._list=a_list
        
    def __getitem__(self, key):
        print("key:", key)
        if isinstance(key, tuple):
            return [self._list[k] for k in key] # represents our tuple
        return self._list[key]
```

```python
obj=my_list([5,5,5])
obj[1]
```

```python
obj[0,1,2]
```
