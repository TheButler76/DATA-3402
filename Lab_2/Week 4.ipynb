### Pseudo Code
#### Creating an A.I.
> #### Create a means of rating the moves
> #### args(board, who will win the game, all possible moves in the current iteration, maximum depth(recursion?))
> #### Should return a list of move and score board.

#### Algorithm Logic
> #### If the current depth is 0, we have reached the bounds of the board and can no longer move the piece. Then add a score to the board.
> #### Find all pieces for the current player.
>> ##### For each piece, consider all possible moves. (Left or Right).
>> ##### Check if the move is possible, then keep track of the move (player_location, direction[Left,Right]).
>> ##### Consider sub-sequence moves by recursively calling.
> ### return a list of moves and scores.

```python
# Heart of the A.I. player
def generate_move(board, player, current_player, depth, keep_all_moves=False):
    if depth == 0:
        return list(), score_board(board, player)
    
    # For each move and score, we have a list.
    moves=list()
    scores=list()
    
    # look through the board for pieces belonging to a specific player.(ixj)
    for i in range(size):
        for j in range(size):
            if board[i][j]==current_player:
                
                # For each piece, consider moving left or right.
                for move in [left_move,right_move]:
                    
                    # Create a copy of the board.
                    new_board=copy.deepcopy(board)
                
                    # Check if it is a valid move.
                    if move_piece(new_board, current_player, (i,j), move, verbose=False):
                        
                        # If so, generate subsequent moves.
                        next_move,score = generate_move(new_board, player,
                                                         switch_player(current_player), depth-1,
                                                         keep_all_moves=keep_all_moves)
                        # Keep track of moves.
                        this_move=[(current_player, (i,j), move)]
                        
                        if keep_all_moves:
                            # We usually need to keep the top moves.
                            this_move.extend(next_move)
                        moves.append(this_move)
                        scores.append(score)
                        
    return moves, scores
```

#### Selecting a move

```python
# Here is a recursive implementation of the algorithm.
def tree_search_0(t, depth=1):
    if isinstance(t[0], list):
        return sum([tree_search(item, depth+1) / depth for item in t])
    else:
        return sum(t)
    
    
def tree_search(t, depth=1):
    if isinstance(t[0], list):
        return sum([tree_search(item, depth+1) for item in t]) / depth
    else:
        return max(t)+min(t)
```

```python
def pick_move(board, player, depth=5, func=max):
    moves, scores = generate_move(board, player,player, depth)
    result=list(map(tree_search, scores))
    move_index=result.index(func(result))
    return moves[move_index][0]
```
