```python
# Find the variable names, proff gave them to us tho.
VarNames = ['signal', 'l_1_pT', 'l_1_eta','l_1_phi',
            'l_2_pT', 'l_2_eta', 'l_2_phi',
           "MET", "MET_phi", "MET_rel", "axial_MET", "M_R",
           "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]
RawNames = [ 'l_1_pT', 'l_1_eta','l_1_phi',
            'l_2_pT', 'l_2_eta', 'l_2_phi',
           "MET", "MET_phi"]
FeatureNames = list(set(VarNames[1:]).difference(RawNames))
FeatureNames
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
filename='SUSY-small.csv'
df = pd.read_csv(filename, dtype='float64', names=VarNames) 
```

```python
df
```

```python
# We are about to use one-hot encoding, I think.
# We have to target the signal col, which already has values between (0-1).
# WE CAN USE A COL AS A METHOD GHASHSFSAFSF, useful. (The signal col is being targeted).

df_sig = df[df.signal==1]
df_bkg = df[df.signal==0]
```

```python
df_sig.shape # tells us the shape of the group
```

```python
df_bkg.shape # tells us the shape of the group
```

```python
# Use a histogram to plot all the columns compared against signal. So exclude the first col [1:]
for col in VarNames[1:]: # staring from list[1]
    print(col)
    plt.figure(figsize=(10,5))
    plt.hist(np.array(df_sig[col]), bins=100, histtype='step', color='red', label="signal",
             density=1, stacked=True)
    plt.hist(np.array(df_bkg[col]), bins=100, histtype='step', color='blue', label="background",
             density=1, stacked=True)
    plt.legend(loc="upper right")
    plt.show()
    
    
```

```python
# We are goiong to use a Monte Carlo simulation on a col
import math # need math to do this problem.
n_bkg_expected = 100
n_sig_expected = 10

# mix signal and background events:
data = df_sig[:n_sig_expected]['M_TR_2'].to_list() + df_bkg[:n_bkg_expected]['M_TR_2'].to_list()
# all of the rows until the number of the signal expected. We compare this to a target col, 'M_TR_2'

# summary: sig is the foreground, bkg is the background, all in a var called data

c, bin_edges = np.histogram(data, bins=20)
# c = num of events in each bin.
# bin_edges is the edge of the bin. WE should know that each bin is follwed by a Poisson distribution.
error = np.sqrt(c)
# error satisfies the Poisson distribution.

bin_width = bin_edges[1]-bin_edges[0] # plotting
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:1]) # plotting

plt.errorbar(bin_centers, c, yerr=error, marker="o") # plotting errorsszzz.

plt.show()
# each veritcal line is the 'error'.
```

```python
# Next, we make a hist for background.
c_bkg_sim, sim_bin = np.histogram(df_bkg['M_TR_2'], bins=100)
_=plt.step(sim_bin[1:], c_bkg_sim) # [1th: to the end]
```

#### y-axis needs to be skewed, so that the area under the curve is 1.
#### SO, create a Probability Density Function 
>#### (F(X) = integral(f(x) dx))

```python
# compute the bin size
bin_size_sim=sim_bin[1]-sim_bin[0]
```

```python
# compute the integral
integral = sum(c_bkg_sim)
```

```python
# now we can skew the yaxis
# divide the counts by the integral, and multiply by expected val

c_bkg_sim_normalize=(n_bkg_expected / integral) * c_bkg_sim
plt.step(sim_bin[1:], c_bkg_sim_normalize)
print("Num of Bkg Expected: ", n_bkg_expected)
print("Integral: ", integral)
print("Bin Size: ", bin_size_sim)
print("Normalized Integral: ", sum(c_bkg_sim_normalize))

# Features used: M_TR_2
```

```python
# We can repeat for signal as well.
c_sig_sim, sim_bin = np.histogram(df_sig['M_TR_2'], bins=100)
integral = sum(c_sig_sim)
c_sig_sim_normalize=(n_sig_expected / integral) * c_sig_sim
plt.step(sim_bin[1:], c_sig_sim_normalize)
```

```python
# plotting error bars
plt.errorbar(bin_centers,c,yerr=error, marker="o", label="data")
plt.yscale("log") # log transformation.
plt.step(sim_bin[1:], c_bkg_sim_normalize, label="background")
plt.step(sim_bin[1:], c_sig_sim_normalize, label='background+signal')
plt.legend()
plt.show()
```

## Lecture 2 | Calculate TPR/FPR

>#### TPR - True Positive Rate.
>#### FPR - False Positive Rate.
>#### derives from the Confusion Matrix.
> #### We calculate because in the future, we should have a ROC curve - . (TPR/FPR)
### Classification

> #### 1.) Accuracy - The percentage. (TP + TN) / (TP + TN + FP + FN)
> #### 2.) Precision - Based on predicted value. (TP / TP + FP)
> #### 3.) Recall - Based on actual positive. (TP / TP + FN)
> #### 4.) TPR - . (TP / TP + FN)
> #### 5.) FPR - . (FP / FP + TN)

```python
TPR, bins_sig,_=plt.hist(
    df_sig["M_TR_2"], bins=100, histtype='step',
    cumulative=-1, density=True)# cummulative being -1 is important. It means to integrate from the right to the left (upper, lower).
_=plt.xlabel("Threshold")
_=plt.ylabel("Probability")
_=plt.suptitle("Cumulative Density Function (signal)")
_=plt.title("TPR(t) = P(M_TR_2 > t)")
```

```python
FPR, bins_sig,_=plt.hist(
    df_bkg["M_TR_2"], bins=bins_sig, histtype='step',
    cumulative=-1, density=True) # cummulative being -1 is important. It means to integrate from the right to the left (upper, lower).
_=plt.xlabel("Threshold")
_=plt.ylabel("Probability")
_=plt.suptitle("Cumulative Density Function (background)") # F(a) - F(b)
_=plt.title("FPR(t) = P(M_TR_2 > t)")
```

```python
plt.plot(FPR,TPR)
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Signal vs Background Discrimination using M_TR_2")
plt.show()
```

## Compute significance. 

```python
n_bkg_expected = 10000
n_sig_expected = 1000
n_sig_expected_prime=n_sig_expected*TPR
n_bkg_expected_prime=n_bkg_expected*FPR

sig=n_sig_expected_prime/np.sqrt(n_sig_expected_prime+n_bkg_expected_prime)
plt.step(bins_sig[:-1], sig)

print("""Significance. 
To the left of the vertex, is the low threshold. 
To the right, the high threshold.
Vertex represents the middle region. Or, the maximum statistical significance.""")
```
