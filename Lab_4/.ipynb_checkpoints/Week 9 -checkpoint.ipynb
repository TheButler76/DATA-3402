### splitting hists into two groups

```python
# We select patients with a cp and those without
cp_0_selection = data['cp']==0
cp_not0_selection = data['cp']!=0
```

```python
# This time we have two variables, cp(with our two masks) and chol
_=plt.hist(data[cp_0_selection]['chol'], bins=50)
_=plt.hist(data[cp_not0_selection]['chol'], bins=50)
# Our goal is to compare two populations.
```

```python
# This time we have two variables, cp(with our two masks) and chol
_=plt.hist(data[cp_0_selection]['chol'], bins=50, label="cp 0")
_=plt.hist(data[cp_not0_selection]['chol'], bins=50, label="cp not 0")
_=plt.legend()
# Our goal is to compare two populations.

# Observations are not easy because the orange overlays the blue.
```

```python
# This time we have two variables, cp(with our two masks) and chol
_=plt.hist(
    data[cp_0_selection]['chol'], bins=50,
    alpha=0.5, color='blue', label="cp 0")
_=plt.hist(
    data[cp_not0_selection]['chol'], bins=50,
    alpha=0.5, color='red', label="cp not 0")
_=plt.legend()
# Our goal is to compare two populations.

# This time we lower the density of the cp not, and made different colors.
# For these two plots, the distribution is not the same, so they are still two different 'distributions'.
```

```python
# to deal with range issues, we standardize and normalize our data/histogram.
_=plt.hist(
    data[cp_0_selection]['chol'], bins=50,
    alpha=0.5, color='blue',
    density=1, label="cp 0")
_=plt.hist(
    data[cp_not0_selection]['chol'], bins=50,
    alpha=0.5, color='red',
    density=1, label="cp not 0")
_=plt.legend()
# with this, they are standardized.
# want to list all the features, so we can make a function to accomplish this.
```

```python
def compare_distributions(data, column_name, selections, **kwargs):
    for label, selection in selections.items():
        _=plt.hist(data[selection][column_name], label=label, **kwargs)
        _=plt.legend()
```

```python
compare_distributions(data, 'chol', 
                      {'cp 0': data['cp']==0,
                      #"cp 1": data['cp']==1.,
                      #"cp 2": data['cp']==2.,
                      #"cp 3": data['cp']==3.,
                      #"cp 4": data['cp']==4.,
                      'cp not 0': data['cp']!=0},
                      bins=25,
                      alpha=0.5,
                      density=1)
# we can use an entire dictionary as an arguement! Useful.
```

```python
compare_distributions(data, 'chol', 
                      {'cp 0': data['cp']==0,
                      "cp 1": data['cp']==1.,
                      "cp 2": data['cp']==2.,
                      "cp 3": data['cp']==3.,
                      "cp 4": data['cp']==4.,
                      'cp not 0': data['cp']!=0},
                      bins=25,
                      alpha=0.5,
                      density=1)
# As you can see here, we can compare many groups.
```

```python
# So now, we can have a histogram for every feature
selection_dict={"cp 0": data['cp']==0,
               "cp not 0": data['cp']!=0,
               }
for column_name in data.columns:
    print(column_name)
    compare_distributions(data, column_name, selection_dict, alpha=0.5, density=1, bins=25)
    
    plt.show()
# This is ok, but we can use a subplot to show all the histograms together.
```

```python
selection_dict={"cp 0": data['cp']==0,
               "cp not 0": data['cp']!=0,
               }

# use subplots to stitch figures together.
# Each plot must be assigned a number.
plt.figure(figsize=(10,10))
for i, column_name in enumerate(data.columns):
    plt.subplot(5,3, i+1) # rows, cols, i plus one if we have more
    compare_distributions(data, column_name, selection_dict,
                          alpha=0.5, density=1, bins=25,
                          stacked=True)
    
    plt.xlabel(column_name)
plt.show()
# It is more h u m a n  r e a d a b l e
```

```python
def compare_cat(data, column_name, selections, **kwargs):
    out=list()
    for label, selection in selections.items():
        total = np.sum(selection)
        data_0 = data[selection]
        values = data_0[column_name].unique()
        out.append([label] + [np.sum(data_0[column_name]==val)/total for val in values])
        
    return out, values
```

```python
# library - 
from IPython.display import HTML, display
from tabulate import tabulate
```

```python
for cat_feature in [data.columns]:
    print(cat_feature)
    results=compare_cat(data, cat_feature, selection_dict)
    print(results)
    display(HTML(tabulate, tabulate(results[0], tablefmt='html')))
```
